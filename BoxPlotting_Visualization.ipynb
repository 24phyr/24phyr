# Customizing the Boxplot

```python
np.random.seed(42)
data1 = np.random.normal(100,20,200)
data2 = np.random.normal(90,10,200)
data3 = np.random.normal(80,30,200)
data4 = np.random.normal(70,40,200)

data = [data1, data2, data3, data4]

plt.figure(figsize=(6,3))
plt.boxplot(data)
plt.show()
```

# Visualizing a standard dataset

```python
dataset = pd.read_csv("Salary_Data.csv")

column = [dataset['Salary'], dataset['YearsExperience']]

plt.boxplot(column)
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
features_to_scale = dataset[['Salary', 'YearsExperience']]
scaled_features = scaler.fit_transform(features_to_scale)
scaled_df = pd.DataFrame(scaled_features, columns=['Scaled_Salary', 'Scaled_YearsExperience'])
plt.boxplot(scaled_df)
plt.show()
```
